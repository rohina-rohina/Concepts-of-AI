# 3.2 Exercises - Data Cleaning and Transformations with ”Titanic Dataset”:
Dataset Used: "titanic.csv"
Problem - 1:
Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
median, maximum value, and minimum value of the ’Fare’ column?

```python
titanic_data=pd.read_csv('Titanic-Dataset.csv')
```

```python
subset_columns=['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
```

```python
titanic_subset= titanic_data[subset_columns]
```

```python
first_class_passengers = titanic_subset[titanic_subset['Pclass'] == 1]
```

```python
fare_mean = first_class_passengers['Fare'].mean()
fare_median = first_class_passengers['Fare'].median()
fare_max = first_class_passengers['Fare'].max()
fare_min = first_class_passengers['Fare'].min()
```

```python
print("Fare Column Statistics for First-Class Passengers:")
print(f"Mean: {fare_mean}")
print(f"Median: {fare_median}")
print(f"Maximum Value: {fare_max}")
print(f"Minimum Value: {fare_min}")
```

How many null values are contained in the ’Age’ column in your subsetted DataFrame? Once you’ve found

```python
null_age_count = first_class_passengers['Age'].isnull().sum()
print(f"Number of null values in 'Age' column: {null_age_count}")
```

```python
import matplotlib.pyplot as plt

mean_survival_rates = titanic_data.groupby('Sex')['Survived'].mean()
print("Mean Survival Rates by Gender:")
print(mean_survival_rates)

genders = mean_survival_rates.index
survival_rates = mean_survival_rates.values

plt.figure(figsize=(8, 5))
plt.bar(genders, survival_rates, color=['orange', 'blue'], alpha=0.7)
plt.title("Mean Survival Rates by Gender", fontsize=16)
plt.ylabel("Mean Survival Rate", fontsize=12)
plt.xlabel("Gender", fontsize=12)
plt.ylim(0, 1)  # Survival rates are proportions (0 to 1)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
survival_by_embarked_and_sex = titanic_data.groupby(['Embarked', 'Sex'])['Survived'].mean().unstack()
print("Mean Survival Rates by Gender and Port of Embarkation:")
print(survival_by_embarked_and_sex)

embarked_ports = survival_by_embarked_and_sex.index  # 'C', 'Q', 'S'
male_survival = survival_by_embarked_and_sex['male']
female_survival = survival_by_embarked_and_sex['female']

x = range(len(embarked_ports))  # X positions for groups

plt.figure(figsize=(10, 6))

# Bar width for grouped bars
bar_width = 0.35

# Bars for males and females
plt.bar(x, male_survival, width=bar_width, label='Male', color='blue', alpha=0.7)
plt.bar([pos + bar_width for pos in x], female_survival, width=bar_width, label='Female', color='orange', alpha=0.7)

# Adding labels and title
plt.xticks([pos + bar_width / 2 for pos in x], embarked_ports)
plt.title("Survival Rates by Gender and Port of Embarkation", fontsize=16)
plt.xlabel("Port of Embarkation", fontsize=12)
plt.ylabel("Mean Survival Rate", fontsize=12)
plt.ylim(0, 1)  # Survival rates are proportions (0 to 1)
plt.legend(title="Gender")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show plot
plt.tight_layout()
plt.show()
```

```python
titanic_data['AgeGroup'] = pd.qcut(titanic_data['Age'].dropna(), 5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'])

survival_by_age_and_class = titanic_data.groupby(['Pclass', 'AgeGroup'])['Survived'].mean().unstack()
print("Mean Survival Rates by Age Group and Pclass:")
print(survival_by_age_and_class)

plt.figure(figsize=(10, 7))

# Plot survival rates for each Pclass
for pclass in survival_by_age_and_class.index:
    plt.plot(
        survival_by_age_and_class.columns,
        survival_by_age_and_class.loc[pclass],
        marker='o',
        label=f'Class {pclass}'
    )

# Customize plot
plt.title("Survival Rates by Age Group and Passenger Class", fontsize=16)
plt.xlabel("Age Group (Quantiles)", fontsize=12)
plt.ylabel("Mean Survival Rate", fontsize=12)
plt.ylim(0, 1)  # Survival rates are proportions (0 to 1)
plt.legend(title="Passenger Class", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show plot
plt.tight_layout()
plt.show()
```
