```python
total_fare = df['Fare'].sum()
```

```python
first_class_fare = df[df['Pclass'] == 1]['Fare'].sum()
```

```python
second_class_fare = df[df['Pclass'] == 2]['Fare'].sum()
```

```python
third_class_fare = df[df['Pclass'] == 3]['Fare'].sum()
```

```python
fare_totals = [first_class_fare, second_class_fare, third_class_fare]
```

```python
fare_proportions = [fare / total_fare for fare in fare_totals]
```

```python
fare_percentages = [proportion * 100 for proportion in fare_proportions]
```

```python
print(f'Percentage of fare revenue by class:')
print(f'First class: {fare_percentages[0]:.2f}%')
print(f'Second class: {fare_percentages[1]:.2f}%')
print(f'Third class: {fare_percentages[2]:.2f}%')
```

```python
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

df['Age_group'] = df['Age'].apply(categorize_age)

total_passengers = len(df)

age_group_counts = df['Age_group'].value_counts()

age_group_proportions = age_group_counts / total_passengers

age_group_percentages = age_group_proportions * 100
print('Percentage of passengers by age group:')
print(age_group_percentages)
```
